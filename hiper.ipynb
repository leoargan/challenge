```python
import requests
import json
import csv

response = requests.get("https://www.hiperlibertad.com.ar/api/catalog_system/pub/products/search/QUESOS%20Y%20FIAMBRES/Quesos/?&_from=0&_to=49&ft&sc=16").text #URL DE LA QUE TOMA LOS DATOS

objeto = json.loads(response) #LEVANTA EL JSON PARA LEER LOS DATOS

```

```python

#CREO LAS LISTAS, UNA DE LOS PRODUCTOS Y LA OTRA DE PRECIOS 
productos = []
for item in objeto:
    productos.append(item['productName'])
print(productos)
```

```python

precio = []
for item in objeto:
   precio.append(item['items'][0]["sellers"][0]["commertialOffer"]["Price"])
print(precio)

descripcion = []
for item in objeto:
   descripcion.append(item['description'])
print(descripcion)
 
preciosindescuento = []
for item in objeto:
  preciosindescuento.append(item['items'][0]["sellers"][0]["commertialOffer"]["PriceWithoutDiscount"])
(preciosindescuento)

categoria = []
for item in objeto:
   categoria.append(item['categoryId'])
print(categoria)

urlproducto = []
for item in objeto:
   urlproducto.append(item['link'])
print(urlproducto)

sku = []
for item in objeto:
   sku.append(item['items'][0]['itemId'])
print(sku)

stock = []
for item in objeto:
  stock.append(item['items'][0]["sellers"][0]["commertialOffer"]["AvailableQuantity"])
(stock)

```

```python
#COMIENZO A TRABAJAR CON PANDAS 

import pandas as pd

dfprod = pd.DataFrame(productos)
dfsku = pd.DataFrame(sku)
dfurl = pd.DataFrame(urlproducto)
dfcat = pd.DataFrame(categoria)
dfstock = pd.DataFrame(stock)
dfplist = pd.DataFrame(preciosindescuento)
dfprecio = pd.DataFrame(precio)
dfdescrip = pd.DataFrame(descripcion)
```

```python
quesos= pd.concat([dfsku, dfprod, dfplist, dfprecio, dfcat, dfdescrip, dfurl, dfstock], axis=1) #CONCATENO PARA CREAR UN DATAFRAME CON EL PRODUCTO Y EL PRECIO 
```
