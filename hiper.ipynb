```python
import requests
import json
import csv

response = requests.get("https://www.hiperlibertad.com.ar/api/catalog_system/pub/products/search/QUESOS%20Y%20FIAMBRES/Quesos/?&_from=0&_to=49&ft&sc=16").text #URL DE LA QUE TOMA LOS DATOS

objeto = json.loads(response) #LEVANTA EL JSON PARA LEER LOS DATOS


#CREO LAS LISTAS, UNA DE LOS PRODUCTOS Y LA OTRA DE PRECIOS 
productos = []
for item in objeto:
    productos.append(item['productName'])
print(productos)

precio = []
for item in objeto:
   precio.append(item['items'][0]["sellers"][0]["commertialOffer"]["Price"])
print(precio)
#revisar 
#preciosindescuento = []
#for item in objeto:
#  precio.append(item['items'][0]["sellers"][0]["commertialOffer"]["PriceWithoutDiscount"])
#print(preciosindescuento)

#datos["ordenes"][0]["toppings"]

#print (productos + precio)
#print (objeto)
```

```python
#COMIENZO A TRABAJAR CON PANDAS 

import pandas as pd

df = pd.DataFrame(productos)
df
```
