```python
import requests
import json
import csv
import re
```

```python
suc = input("ingrese sucursal de 1 a 16: ") #ingreso manual de sucursal 
products = []

for i in range(0, 50):  # iterar 20 veces
    response = requests.get(f"https://www.hiperlibertad.com.ar/api/catalog_system/pub/products/search/?&_from={i*50}&_to={(i+1)*50-1}&ft&sc={suc}").text 
    objeto = json.loads(response)
    products += objeto

# Aquí puedes procesar los productos obtenidos
```
